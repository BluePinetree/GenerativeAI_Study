### 2-3 연속형 데이터의 범주화(BloodPressure)
* `BloodPressure`의 열(column) 값을 4개 구간으로 분류
    - 0~20 : A
    - 21~30 : B
    - 31~40 : C
    - 41~max : D
* 분류된 값들을 `bloodpressure_cut`이라는 열을 생성하여 값을 할당

```python
train['BloodPressure']
```

```python
# BloodPressure열의 모든 원소에 동일하게 적용시킬 함수 정의
def bloodpressure_cut(x):
  result = ''
  if 0 <= x < 20:
    result = 'A'
  elif x < 30:
    result = 'B'
  elif x < 40:
    result = 'C'
  else:
    result = 'D'

  return result
```

```python
# map메서드를 사용하여 범주화된 특징 bloodpressure_cut 특징 정의
train['bloodpressure_cut'] = train['BloodPressure'].map(bloodpressure_cut)
test['bloodpressure_cut'] = test['BloodPressure'].map(bloodpressure_cut)
```

### 2-4. 연속형 데이터의 범주화2(Glucose)
* `Glucose` 피처에 분위수 기준으로 4가지로 분류하기
* ['normal', 'suspected1', 'suspected2', 'danger']

```python
# train['Glucose'] = pd.qcut(train['Glucose'], q = 4, labels=['a','b','c','d'])
# train['Glucose'].value_counts(sort = False)
```

```python
train['Glucose']
```

```python
def glucose_cut(x):
    result = ''
    try:
        x = int(x)
        if x < 99:
          result = 'normal'
        elif x < 117:
          result = 'suspected1'
        elif x < 142:
          result = 'suspected2'
        else:
          result = 'danger'
    except ValueError:
        result = 'invalid'  # 숫자로 변환할 수 없는 경우 처리
    return result
```

```python
# train['Glucose']
```

```python
train['glucose_cut'] = train['Glucose'].map(glucose_cut)
test['glucose_cut'] = test['Glucose'].map(glucose_cut)
```

```python
x_train = train.drop('Outcome', axis = 1)
x_test = test
y_train = train['Outcome']
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(train.loc[:,'Pregnancies':'Age'])
```

```python
scaled_train = scaler.transform(train.loc[:,'Pregnancies':'Age'])
scaled_test = scaler.transform(test.loc[:,'Pregnancies':'Age'])

sequential_features = scaler.feature_names_in_

for index, feature in enumerate(sequential_features):
    train[feature] = scaled_train[:,index]
    test[feature] = scaled_test[:,index]

train.head()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

categorical_data = ['glucose_cut', 'bloodpressure_cut']

for feature in categorical_data:
  le = le.fit(x_train[feature])
  x_train[feature] = le.transform(x_train[feature])

  for label in x_test[feature]:
    if label not in le.classes_:
      le.classes_ = np.append(le.classes_, label)

  x_test[feature] = le.transform(x_test[feature])
```

## 3. 모델링

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC

# 모델 초기화 (max_iter값이 400이상일경우 정확도 비슷),n_estimators값 커지면 과대적합
model_lr = LogisticRegression(max_iter=400)
model_rf = RandomForestClassifier(n_estimators=1)
model_svm = SVC(C=1.0, gamma='scale')

# 모델 학습
model_lr.fit(x_train, y_train)
model_rf.fit(x_train, y_train)
model_svm.fit(x_train, y_train)

# 훈련 데이터에 대한 예측
y_pred_lr = model_lr.predict(x_train)
y_pred_rf = model_rf.predict(x_train)
y_pred_svm = model_svm.predict(x_train)
# 훈련 데이터 정확도 출력
accuracy_train_lr = accuracy_score(y_train, y_pred_lr)
accuracy_train_rf = accuracy_score(y_train, y_pred_rf)
accuracy_train_svm = accuracy_score(y_train, y_pred_svm)
print(f"Training Accuracy (Logistic Regression): {accuracy_train_lr}")
print(f"Training Accuracy (Random Forest): {accuracy_train_rf}")
print(f"Training Accuracy (Support Vector Machine, SVM): {accuracy_train_svm}")

```
