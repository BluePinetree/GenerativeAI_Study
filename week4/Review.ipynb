### 2-3 연속형 데이터의 범주화(BloodPressure)
* `BloodPressure`의 열(column) 값을 4개 구간으로 분류
    - 0~20 : A
    - 21~30 : B
    - 31~40 : C
    - 41~60 : D
* 분류된 값들을 `bloodpressure_cut`이라는 열을 생성하여 값을 할당

```python
def bloodpressure_cut(x):
  result = ''
  if 0 <= x <= 20:
    result = 'A'
  elif x <= 30:
    result = 'B'
  elif x <= 40:
    result = 'C'
  else:
    result = 'D'
  return result
```

```python
train['bloodpressure_cut'] = train['BloodPressure'].map(bloodpressure_cut)
```

```python
train['bloodpressure_cut'].value_counts()
```

```python
test['bloodpressure_cut'] = test['BloodPressure'].map(bloodpressure_cut)
```

### 2-4. 연속형 데이터의 범주화2(Glucose)
* `Glucose` 피처에 분위수 기준으로 4가지로 분류하기
* ['normal', 'suspected1', 'suspected2', 'danger']

```python
train[['Glucose']].describe()
```

```python
def glucose_cut(x):
  result = ''
  if 0 <= x <= 99:
    result = 'normal'
  elif x <= 117:
    result = 'suspected1'
  elif x <= 142:
    result = 'suspected2'
  else:
    result = 'danger'
  return result
```

```python
train['glucose_cut'] = train['Glucose'].map(glucose_cut)
```

```python
train['glucose_cut'].value_counts()
```

```python
test['glucose_cut'] = test['Glucose'].map(glucose_cut)
# train 데이터의 분위수에 따라 분류
```

```python
test['glucose_cut'].value_counts()
```

```python
train['glucose_cut']
```

### 2-5. 표준화 및 인코딩
* 수치형 데이터 : 표준화
* 범주형 데이터 : 레이블인코딩

```python
x_train = train.drop(['Outcome'], axis = 1)
y_train = train['Outcome']
x_test = test
# ID column은 위에서 삭제했음
```

```python
x_train.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(x_train.loc[:,'Pregnancies':'Age'])
```

```python
scaled_train = scaler.transform(x_train.loc[:,'Pregnancies':'Age'])
scaled_test = scaler.transform(x_test.loc[:,'Pregnancies':'Age'])

sequential_features = scaler.feature_names_in_

for index, feature in enumerate(sequential_features):
    x_train[feature] = scaled_train[:,index]
    x_test[feature] = scaled_test[:,index]
```

```python
x_train.head()
```

```python
x_test.head()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
```

```python
categorical_data = ['glucose_cut', 'bloodpressure_cut']

for feature in categorical_data:
  le = le.fit(x_train[feature])
  x_train[feature] = le.transform(x_train[feature])

  for label in x_test[feature]:
    if label not in le.classes_:
      le.classes_ = np.append(le.classes_, label)

  x_test[feature] = le.transform(x_test[feature])
```

```python
x_train.head()
```

## 3. 모델링

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
```

```python
model.fit(x_train, y_train)

predict = model.predict(x_test)
```

```python
sample['Outcome'] = predict
sample.to_csv('submission.csv', index = False)
```
